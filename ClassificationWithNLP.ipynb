PREPROCESSING THE TEXT DATA

```python
nltk.download('stopwords')
nltk.download('wordnet')
lematizer=WordNetLemmatizer()
lb=LabelEncoder()
cv=CountVectorizer(max_features=2000)
print(stopwords.words('english'))
```

```python
def preprocessingText(text):
    # Remove special characters
    text=re.sub("[^a-z A-Z0-9]","",text)
    # Lower case convertion
    text=text.lower()
    return text
```

```python
def preprocessingTextUsingNLTK(text):
    text=preprocessingText(text)
    allEngStopWords=stopwords.words("english")
    #removal of stopwords and Lemmatizing
    if text!=None:
        text= " ".join([ lematizer.lemmatize(word) for word in text.split(" ") if word not in set(allEngStopWords) ])
    else:
        print(text)
    return text
```

```python
def preprocessingTextUsingSPACY(text):
    text=preprocessingText(text)
    lemWithStop=spacy.load("en_core_web_sm")
    doc=lemWithStop(text)
    #removal of stopwords and Lemmatizing
    if doc!=None and len(doc):
        text= " ".join([ word.lemma_ for word in doc if not word.is_stop ])
    else:
        print(text)
    return text
```

Stopword and Lemmatizing either NLTK or SPACY

```python
df['Text']=df['Text'].apply(preprocessingTextUsingSPACY)
# df['Text']=df['Text'].apply(preprocessingTextUsingNLTK)
y=lb.fit_transform(df['Label'])
df.drop(columns=['Label'],inplace=True)
arr=cv.fit_transform(df['Text']).toarray()
X=pd.DataFrame(arr)
X.head()
```

```python
y=y.reshape(-1,1)
y.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=20,test_size=0.2)
gnb=GaussianNB()
gnb.fit(X_train,y_train)
```

BOTH SPACY AND NLTK GAVE SAME ACCURACY OF 81%, Other csv files also work by replacing the file path in the second cell

```python
y_pred=gnb.predict(X_test)
accuracy_score(y_test,y_pred)
```
